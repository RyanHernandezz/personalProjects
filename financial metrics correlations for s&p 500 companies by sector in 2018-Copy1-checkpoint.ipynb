I'm intrested in how the correlation between differnent financial metrics might look in the stock market as a whole and in different sectors of the stock market

I'll be using data from the S&p 500 in 2018 from https://datahub.io/core/s-and-p-500-companies-financials to acomplish this

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading data into dataframe
full_table = pd.read_csv('constituents-financials_csv.csv')
```

```python
full_table.head()
```

for my purposes some of the information here isn't needed

```python
#taking away the listed columns as they aren't needed
full_table = full_table.drop(columns=['Symbol', 'Name', 'Price','52 Week Low', '52 Week High', 'SEC Filings'])
```

It would also be helpfull to take away the spaces in the column names

```python
#taking away spaces in the sector column since they can be an inconvenience
full_table['Sector'] = full_table['Sector'].str.replace(' ', '_')
```

Now I want to take a glance at the correlations between these values

```python
#looking at some correlations between the different values (pearson correlations)
full_table.corr()
```

I'd like to draw a heatmap of these correlations now to better visualize them

```python
#a heatmap makes this a lot easier to look at
#making the correlation table
full_table_corr=full_table.corr()
#making the heatmap
x = sns.heatmap(full_table_corr, xticklabels=full_table_corr.columns, yticklabels=full_table_corr.columns)
#making the title for the heatmap
x.set_title('All'+' Sectors')
```
