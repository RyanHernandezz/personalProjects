```python
import pandas as pd
import numpy as np
import seaborn as sns
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
#loading the data
df = pd.read_csv('medicare_data.csv')
df.head()
```

```python
#taking only 2021 variables
df2021 = df[['Brnd_Name','Gnrc_Name','Tot_Mftr','Mftr_Name','Tot_Spndng_2021', 'Tot_Dsg_Unts_2021', 'Tot_Clms_2021', 'Tot_Benes_2021','Avg_Spnd_Per_Dsg_Unt_Wghtd_2021','Avg_Spnd_Per_Clm_2021','Avg_Spnd_Per_Bene_2021','Chg_Avg_Spnd_Per_Dsg_Unt_20_21','CAGR_Avg_Spnd_Per_Dsg_Unt_17_21']]
df2021.head()
```

```python
df2021 = df2021.sort_values('Tot_Spndng_2021', ascending=True)
df2021 = df2021.round(8)
df2021.tail()
```

```python
spending = df['Tot_Spndng_2021']
spending.sum()
```

```python
manufacturers = df2021['Mftr_Name'].unique()
index = np.argwhere(manufacturers=='Overall')
manufacturers = np.delete(manufacturers, index)
print(manufacturers)
```

```python
final_df = pd.DataFrame(columns=['manufacturers', 'Total_Spending'])
for value in manufacturers:
    rows = df2021[df2021['Mftr_Name'] == value]
    total_spending = rows['Tot_Spndng_2021'].sum()
    final_df = final_df.append({'manufacturers': value, 'Total_Spending': total_spending}, ignore_index=True)

print(final_df)
```

```python
final_df.head()
```

```python
sum_values = final_df['Total_Spending'].sum()

final_df['%_of_total'] = final_df['Total_Spending'] / sum_values
final_df.head(50)
```

```python
spending = final_df['%_of_total']
spending.sum()
```
