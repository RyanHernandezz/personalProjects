```python
import pandas as pd
import csv
from fuzzywuzzy import fuzz, process
import datetime


#define assignment name
assignment_name = 'Quiz Oct 5 (822107)'
#assignment_name = 'Quiz Oct 3 (821631)'

# Define the file path for gradebook
gradebookexport = r'C:\Users\ryan\Downloads\recentexport105b.csv'

# Specify the path to your CSV file for the google sheet inputs
csv_file_path = r'C:\Users\ryan\downloads\participation105.csv'






# Use pandas to read the gradebook CSV file into a DataFrame
gradebook = pd.read_csv(gradebookexport)

# List of names in the format 'First-name Last-name'
converted_names = gradebook['Student'][2:].tolist()
#converted_names = ['Student, Test']
original_names = []

# Iterate through the converted_names list and convert each name back to the original format
for converted_name in converted_names:
    # Remove the brackets and split the name into first name and last name
    first_name, last_name = converted_name.strip('[]').split(', ')
    # Create the original name in the format "First-name Last-name"
    original_name = f"{last_name} {first_name}"
   
    # Check if the original_name is already in the list
    if original_name in original_names:
        print(f"Duplicate name found: {original_name}")
        # Stop the interpreter or raise an exception to halt execution
        raise ValueError("Duplicate names found. Stopping execution.")
       
    # Append the original name to the new list
    original_names.append(original_name)

# Specify the path to your CSV file for the google sheet inputs
#csv_file_path = r'C:\Users\ryan\downloads\participation103.csv'

# Initialize an empty list to store the rows as lists of strings
# Initialize an empty list to store the row as a list of strings
list_of_strings = []

# Open the CSV file and read its contents
with open(csv_file_path, 'r', newline='') as csv_file:
    csv_reader = csv.reader(csv_file)
    # Assuming there is only one row in the CSV
    row = next(csv_reader, None)
    if row:
        # Convert each element in the row to a string
        list_of_strings = [str(item) for item in row]

# Initialize an empty list to store the best-match values
best_matches = []

# Define a threshold for fuzzy matching (adjust as needed)
threshold = 30  # Adjust this threshold based on your matching requirements

for value in list_of_strings:
    # Use process.extractOne to find the best match
    best_match, score = process.extractOne(value, original_names)

    # Check if the match score is above the threshold
    if score >= threshold:
        best_matches.append(best_match)

converted_names = []

# Iterate through the original list and convert each name
for name in best_matches:
    # Split the name into parts based on spaces
    parts = name.split()

    # Check if there are at least two parts (first name and last name)
    if len(parts) >= 2:
        first_name = parts[0]
        last_name = " ".join(parts[1:])  # Combine remaining parts as last name
        # Create the converted name in the format "Last-name, First-name"
        converted_name = f"{last_name}, {first_name}"
        # Append the converted name to the new list
        converted_names.append(converted_name)
    else:
        # Handle names that don't have both first and last names
        converted_names.append(name)  # Append the original name
       
for student_name in converted_names:
    # Check if 'Student' in 'gradebook' matches 'student_name'
    mask = gradebook['Student'] == student_name

    # Update the 'Quiz Sep 7 (804647)' column for matching rows
    #gradebook.loc[mask, 'Quiz Sep 26 (816579)'] = 1  # Set the new score here
    gradebook.loc[mask, assignment_name] = 1  # Set the new score here
# Get the current date and time
current_datetime = datetime.datetime.now()

# Format the date and time as a string
timestamp = current_datetime.strftime('%d-%H-%M')

# Define the file path for the new CSV file
directory = r'C:\Users\ryan\Downloads'
updated_gradebook_export = fr'{directory}\updatedgradebook_{timestamp}.csv'
gradebook.to_csv(updated_gradebook_export, index=False)
print(f"Gradebook saved as {updated_gradebook_export}")
```
